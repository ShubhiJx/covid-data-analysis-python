Visual for repo readme.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# load our dataset
df = pd.read_csv("../data/covid_small.csv")

# countries list
countries = [
    'India',
    'United States'
]

# filter data
df = df[df["location"].isin(countries)]

# convert table format
pivot = df.pivot(index="location", columns="date", values="new_cases")

# fill missing values
pivot = pivot.fillna(0)

# convert to numpy
plot_df = np.sqrt(pivot.to_numpy())

# plot heatmap
plt.figure(figsize=(10,4))

plt.imshow(plot_df, aspect='auto')

plt.yticks(range(len(pivot.index)), pivot.index)

plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=45)

plt.colorbar()

plt.title("COVID New Cases Heatmap")

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
df = pd.read_csv("../data/covid_small.csv")

# filter India data
india = df[df["location"]=="India"].copy()

# moving average (use 3 instead of 7)
india["3day_avg"] = india["new_cases"].rolling(3).mean()

# style
plt.style.use("ggplot")

# plot
plt.figure()

plt.plot(india["date"], india["new_cases"], label="Daily cases")

plt.plot(india["date"], india["3day_avg"], label="3-day average")

plt.legend()

plt.title("India Daily Cases with Moving Average")

plt.xlabel("Date")
plt.ylabel("Cases")

plt.xticks(rotation=45)

plt.show()
```
